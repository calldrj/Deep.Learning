```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
import pandas as pd

# Function plots the loss in training and validation dataset
def plot_loss(df_0, title_0, df_1, title_1):
    fig, axes = plt.subplots(2, figsize=(10,12), sharex=True)
    df_0.plot(kind='line', x='epoch', y='loss_train', ax=axes[0], title=title_0)
    df_0.plot(kind='line', x='epoch', y='loss_validation', ax=axes[0])
    df_1.plot(kind='line', x='epoch', y='loss_train', ax=axes[1], title=title_1)
    df_1.plot(kind='line', x='epoch', y='loss_validation', ax=axes[1])
    
# Function plots the accuracy in training and validation dataset
def plot_accuracy(df_0, title_0, df_1, title_1):
    fig, axes = plt.subplots(2, figsize=(10,12), sharex=True)
    df_0.plot(kind='line', x='epoch', y='accuracy_train', ax=axes[0], title=title_0)
    df_0.plot(kind='line', x='epoch', y='accuracy_validation', ax=axes[0])
    df_1.plot(kind='line', x='epoch', y='accuracy_train', ax=axes[1], title=title_1)
    df_1.plot(kind='line', x='epoch', y='accuracy_validation', ax=axes[1])
    
```

```python
# Read the data of 400 epochs x 1k samples runs
# for MSE and Cross-Entrophy losses
df_50K = pd.read_csv("CE[75_10_0.005_0.005_1].csv")
df_10K = pd.read_csv("CE[75_10_0.005_0.005_5].csv")

df_50K.head()
```

```python
df_10K.head()
```

```python
plot_loss(df_10K, "Training Size: 10k", df_50K, "Training Size: 50k" )
```

```python
plot_accuracy(df_10K, "Training Size: 10k", df_50K, "Training Size: 50k" )
```
