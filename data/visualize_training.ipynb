```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
import pandas as pd

# Function plots the loss in training and validation dataset
def plot_loss(df_0, df_1):
    fig, axes = plt.subplots(2, figsize=(10,12), sharex=True)
    df_0.plot(kind='line', x='epoch', y='loss_train', ax=axes[0], 
              title="Loss Function: Mean Squared Error")
    df_0.plot(kind='line', x='epoch', y='loss_validation', ax=axes[0])
    df_1.plot(kind='line', x='epoch', y='loss_train', ax=axes[1], 
              title="Loss Function: Cross Entrophy")
    df_1.plot(kind='line', x='epoch', y='loss_validation', ax=axes[1])
    
# Function plots the accuracy in training and validation dataset
def plot_accuracy(df_0, df_1):
    fig, axes = plt.subplots(2, figsize=(10,12), sharex=True)
    df_0.plot(kind='line', x='epoch', y='accuracy_train', ax=axes[0], 
              title="Loss Function: Mean Squared Error")
    df_0.plot(kind='line', x='epoch', y='accuracy_validation', ax=axes[0])
    df_1.plot(kind='line', x='epoch', y='accuracy_train', ax=axes[1], 
              title="Loss Function: Cross Entrophy")
    df_1.plot(kind='line', x='epoch', y='accuracy_validation', ax=axes[1])
    
```

```python
# Read the data of 400 epochs x 1k samples runs
# for MSE and Cross-Entrophy losses
df_400_MSE = pd.read_csv("neuron_net_400_MSE.csv")
df_400_CE = pd.read_csv("neuron_net_400_CE.csv")

df_400_MSE.head()
```

```python
df_400_CE.head()
```

```python
plot_loss(df_400_MSE, df_400_CE)
```

```python
plot_accuracy(df_400_MSE, df_400_CE)
```

```python
# Read the data of 30 epochs x 50k samples runs 
# for MSE and Cross-Entrophy losses
df_30_MSE = pd.read_csv("neuron_net_30_MSE.csv")
df_30_CE  = pd.read_csv("neuron_net_30_CE.csv")
```

```python
plot_loss(df_30_MSE, df_30_CE)
```

```python
plot_accuracy(df_30_MSE, df_30_CE)
```
