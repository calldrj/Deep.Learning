```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
import pandas as pd
df = pd.read_csv("neuron_net.csv")
```

```python
df.head()
```

```python
ax1 = plt.gca()
df.plot(kind='line', x='epoch', y='loss_train', ax=ax1)
df.plot(kind='line', x='epoch', y='loss_validation', ax=ax1)
plt.show()
```

```python
ax2 = plt.gca()
df.plot(kind='line', x='epoch', y='accuracy_train', ax=ax2)
df.plot(kind='line', x='epoch', y='accuracy_validation', ax=ax2)
plt.show()
```
